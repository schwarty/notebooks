```python
import os
import json
import glob

import numpy as np
import pandas as pd
import pylab as pl

from sklearn.metrics import precision_recall_fscore_support
```

```python
metrics = ['precision', 'recall', 'fscore', 'support']

def classification_scores_frame(data_frame, labels=None):
    if labels is None:
        labels = data_frame.y_true.unique()
    scores = dict(zip(metrics, precision_recall_fscore_support(data_frame.y_true, data_frame.y_pred, labels=labels)))
    scores['label'] = labels
    return pd.DataFrame(scores, index=labels)
```
