```python
import os
import json
import glob

import numpy as np
import pandas as pd
import pylab as pl

from sklearn.metrics import precision_recall_fscore_support
```

```python
metrics = ['precision', 'recall', 'fscore', 'support']

with open('/havoc/gridly/labels.json') as f:
    labels = json.load(f)
    
scores_frame = pd.DataFrame.from_csv('/havoc/gridly/scores.csv')
```

```python
scores_frame.head()
```

```python
pos_label_frame = scores_frame[scores_frame['class'] == 1]
```

```python
pos_label_frame.head(10)
```

```python
reduce_frame = pos_label_frame.groupby(['label', 'set_type', 'C'])[metrics].agg([np.mean, np.std])
```

```python
reduce_frame.head(20)
```

```python
for (label, mean_frame), (label, std_frame) in zip(reduce_frame[('fscore', 'mean')].groupby(level='label'), reduce_frame[('fscore', 'std')].groupby(level='label')):
    for (frame_id, fscore_frame), (frame_id, std_fscore_frame) in zip(mean_frame.groupby(level='set_type'), std_frame.groupby(level='set_type')):
        yerr = np.hstack([std_fscore_frame.values[:, None], -std_fscore_frame.values[:, None]])
        y = fscore_frame.values
        color = 'b' if frame_id == 'test' else 'g'
        ax = fscore_frame.plot(kind='line', label=frame_id, figsize=(6, 4), ylim=(0, 1))
        ax.fill_between(range(len(fscore_frame)), y+yerr[:, 0], y+yerr[:, 1], alpha=.5, color=color)
        ax.set_xticks(range(len(fscore_frame)));
        ax.set_xticklabels([i[-1] for i in fscore_frame.index]);
        ax.set_xlabel('C');
        ax.set_ylabel('f1-score');
        legend(loc='best', ncol=3)
    title('%s' % (label))
    show()
```
